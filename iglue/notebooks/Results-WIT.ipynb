```python
import pandas as pd
from collections import defaultdict
```

```python
model2name = {
    "ctrl_muniter": "mUNITER",
    "ctrl_xuniter": "xUNITER",
    "uc2": "UC$^2$",
    "m3p": "M$^3$P",
}
```

```python
dset = 'WIT'
langs = ['ar', 'bg', 'da', 'el', 'et', 'id', 'ja', 'ko', 'tr', 'vi']
```

```python
dset_0 = pd.read_csv(f"../results/{dset.lower()}/{dset}_ir_0.csv")
dset_0
```

```python
print('\multirow{4}{*}{\small \shortstack{IR}}', end=' ')
for _, row in dset_0.iterrows():
    print(f"& {model2name[row['model']]}", end=' ')
    for l in langs:
        print('& %.2f' % row[l], end=' ')
    print('\\\\')
```

## IR

```python
lang2model2vals = {lang: {m: [] for m in model2name} for lang in langs}

dset_0 = pd.read_csv(f"../results/{dset.lower()}/{dset}_ir_0.csv")
for lang in lang2model2vals:
    if lang not in dset_0.columns:
        lang2model2vals[lang][m].append('N/A')
        continue
    for (m, v) in dset_0[['model', lang]].values:
        lang2model2vals[lang][m].append(v)
        
for s in shots:
    dset_0 = pd.read_csv(f"../results/{dset.lower()}/{dset}_ir_{s}.csv")
    for lang in lang2model2vals:
        if lang not in dset_0.columns:
            lang2model2vals[lang][m].append('-')
            continue
        for (m, v) in dset_0[['model', lang]].values:
            lang2model2vals[lang][m].append(v)
```

```python
for lang in lang2model2vals:
    print('\multirow{4}{*}{\small \shortstack{%s}}' % lang, end=' ')
    model2vals = lang2model2vals[lang]
    for m, v in model2vals.items():
        print(f'& {model2name[m]}', end=' ')
        for e in v:
            print('& %.2f' % e, end=' ')
        print('\\\\')
    print('\midrule')
```
