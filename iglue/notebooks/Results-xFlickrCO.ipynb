```python
import glob
import numpy as np
import pandas as pd
from collections import defaultdict
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'Times New Roman'

import matplotlib
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.preview'] = True
plt.rc('font', family='serif', serif=['Times'])
import warnings
warnings.filterwarnings("ignore")
```

```python
model2name = {
    "ctrl_muniter": "mUNITER",
    "ctrl_xuniter": "xUNITER",
    "uc2": "UC$^2$",
    "m3p": "M$^3$P",
}
```

```python
dset = 'xFlickrCO'
langs = ['de', 'es', 'id', 'ja', 'ru', 'tr', 'zh']
shots = [1,5,10,25,50,100]
```

## IR

```python
lang2model2vals = {lang: {m: [] for m in model2name} for lang in langs}

dset_0 = pd.read_csv(f"../results/{dset.lower()}/{dset}_ir_0.csv")
for lang in lang2model2vals:
    if lang not in dset_0.columns:
        lang2model2vals[lang][m].append('N/A')
        continue
    for (m, v) in dset_0[['model', lang]].values:
        lang2model2vals[lang][m].append(v)
        
for s in shots:
    dset_0 = pd.read_csv(f"../results/{dset.lower()}/{dset}_ir_{s}.csv")
    for lang in lang2model2vals:
        if lang not in dset_0.columns:
            lang2model2vals[lang][m].append('-')
            continue
        for (m, v) in dset_0[['model', lang]].values:
            lang2model2vals[lang][m].append(v)
```

```python
for lang in lang2model2vals:
    print('\multirow{4}{*}{\small \shortstack{%s}}' % lang, end=' ')
    model2vals = lang2model2vals[lang]
    for m, v in model2vals.items():
        print(f'& {model2name[m]}', end=' ')
        for e in v:
            print('& %.2f' % e, end=' ')
        print('\\\\')
    print('\midrule')
```

## IR Performance by xFlickr subset

### English

```python
import json
import numpy as np

dset = "xFlickrCO"
sub_models = ['ctrl_muniter', 'ctrl_xuniter', 'uc2', 'm3p']
langs = ['en']

model2flickr_en = {}
model2coco_en = {}

# zero-shot
for m in sub_models:
    m_flickr_accs = []
    m_coco_accs = []
    for lang in langs:
        j = json.load(open(f"../results/{dset.lower()}/zero_shot/{m}/test_{lang}_0_result.json"))
        res = []
        for ix, l in enumerate(j):
            res.append(l[0] == ix)
        flickr_acc = np.sum(res[:1000])/1000
        coco_acc = np.sum(res[1000:])/1000
        m_flickr_accs.append(flickr_acc)
        m_coco_accs.append(coco_acc)
    model2flickr_en[m] = np.mean(m_flickr_accs)
    model2coco_en[m] = np.mean(m_coco_accs)
    
print(model2flickr_en)
print(model2coco_en)
```

### Multilingual

```python
import json
import numpy as np

dset = "xFlickrCO"
sub_models = ['ctrl_muniter', 'ctrl_xuniter', 'uc2', 'm3p']
langs = ['de', 'es', 'id', 'ja', 'ru', 'tr', 'zh']

model2flickr_avgs = defaultdict(list)
model2coco_avgs = defaultdict(list)
shots = []

# zero-shot
for m in sub_models:
    m_flickr_accs = []
    m_coco_accs = []
    for lang in langs:
        j = json.load(open(f"../results/{dset.lower()}/zero_shot/{m}/test_{lang}_0_result.json"))
        res = []
        for ix, l in enumerate(j):
            res.append(l[0] == ix)
        flickr_acc = np.sum(res[:1000])/1000
        coco_acc = np.sum(res[1000:])/1000
        m_flickr_accs.append(flickr_acc)
        m_coco_accs.append(coco_acc)
    model2flickr_avgs[m].append(np.mean(m_flickr_accs))
    model2coco_avgs[m].append(np.mean(m_coco_accs))
shots.append(0)

# few-shot
for s in [1,5,10,25,50,100]:
    shots.append(s)
    for m in sub_models:
        m_flickr_accs = []
        m_coco_accs = []
        for lang in langs:
            j = json.load(open(f"../results/{dset.lower()}/few_shot/{m}/test_{lang}_{s}_result.json"))
            res = []
            for ix, l in enumerate(j):
                res.append(l[0] == ix)
            flickr_acc = np.sum(res[:1000])/1000
            coco_acc = np.sum(res[1000:])/1000
            m_flickr_accs.append(flickr_acc)
            m_coco_accs.append(coco_acc)
        model2flickr_avgs[m].append(np.mean(m_flickr_accs))
        model2coco_avgs[m].append(np.mean(m_coco_accs))
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'Times New Roman'

colors = ['#000000', '#E69F00', '#56B4E9', '#009E73', '#CC79A7']
markers = ['X', 's', '^', 'o', 'd']

f, axs = plt.subplots(1, 2, figsize=(16,6), sharey=True)

im = 1
for m, ys in model2flickr_avgs.items():
    xs = np.array(shots)
    ys = 100*np.array(ys)
    axs[0].plot(xs, ys, label=model2name[m], marker=markers[im], markersize=15, linewidth=4, markeredgecolor='k', color=colors[im])
    im += 1
    
im = 1
for m, ys in model2coco_avgs.items():
    xs = np.array(shots)
    ys = 100*np.array(ys)
    axs[1].plot(xs, ys, label=model2name[m], marker=markers[im], markersize=15, linewidth=4, markeredgecolor='k', color=colors[im])
    im += 1

axs[0].set_title('Flickr subset', fontsize=36)
axs[0].set_ylim(0,22)
axs[0].set_ylabel('Accuracy', fontsize=36)
axs[1].set_title('COCO subset', fontsize=36)
axs[1].set_ylim(0,22)

for ix, ax in enumerate(axs):
    ax.grid(alpha=0.3)
    ax.tick_params(axis='both', which='major', labelsize=24)
    ax.margins(x=0.05)
    ax.set_xticks([0,1,5,10,25,50,100])
    ax.set_xticklabels(['',1,5,10,25,50,100])
axs[0].legend(loc='lower right', ncol=2, title='\\textbf{Model}', title_fontsize=26, fontsize=22)

for ax in axs:
    ax.axvline(0, linestyle='-', linewidth=2, color='gray', alpha=0.5)

f.tight_layout(pad=2.5)
f.savefig("fewshot_curves_xFlickrCO-subset.pdf", bbox_inches='tight')
```
